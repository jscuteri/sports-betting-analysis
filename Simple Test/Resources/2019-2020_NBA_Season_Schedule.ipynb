```python
# Import Dependencies

import pandas as pd
import numpy as np
import os
```

```python
# Load Schedule Data

october_2019 = "2019-2020 Schedule Results/October_2019.csv"
november_2019 = "2019-2020 Schedule Results/November_2019.csv"
december_2019 = "2019-2020 Schedule Results/December_2019.csv"
january_2020 = "2019-2020 Schedule Results/January_2020.csv"
february_2020 = "2019-2020 Schedule Results/February_2020.csv"
march_2020 = "2019-2020 Schedule Results/March_2020.csv"
july_2020 = "2019-2020 Schedule Results/July_2020.csv"
august_2020 = "2019-2020 Schedule Results/August_2020.csv"
september_2020 = "2019-2020 Schedule Results/September_2020.csv"
october_2020 = "2019-2020 Schedule Results/October_2020.csv"
```

```python
# Read each data set

october19_schedule = pd.read_csv(october_2019)
november19_schedule = pd.read_csv(november_2019)
december19_schedule = pd.read_csv(december_2019)
january20_schedule = pd.read_csv(january_2020)
february20_schedule = pd.read_csv(february_2020)
march20_schedule = pd.read_csv(march_2020)
july20_schedule = pd.read_csv(july_2020)
august20_schedule = pd.read_csv(august_2020)
september20_schedule = pd.read_csv(september_2020)
october20_schedule = pd.read_csv(october_2020)
```

```python
# Append the 2019 datasets into one large one

df = october19_schedule.append(november19_schedule, ignore_index=True)
df = df.append(december19_schedule, ignore_index=True)
df = df.append(january20_schedule, ignore_index=True)
df = df.append(february20_schedule, ignore_index=True)
df = df.append(march20_schedule, ignore_index=True)
df = df.append(july20_schedule, ignore_index=True)
df = df.append(august20_schedule, ignore_index=True)
df = df.append(september20_schedule, ignore_index=True)
df = df.append(october20_schedule, ignore_index=True)
df.head()
```

```python
# Remove unwanted colummns from each

df.drop(['Start (ET)', 'Unnamed: 6', 'Unnamed: 7', "Attend.", "Notes"], axis=1, inplace=True)
df
```

```python
# Rename columns

df.rename(columns={'Visitor/Neutral': 'Away', 'Home/Neutral': 'Home', 'PTS.1':'Home Score' , 'PTS' : 'Away Score'}, inplace=True)
df
```

```python
# Add a win column for the home team as a numeric value

home_winning_condition = [
    (df['Home Score'] > df['Away Score'])
]

home_win_value = [1]

df['Home Win'] = np.select(home_winning_condition, home_win_value)
```

```python
# Add a margin of victory column

df['Winning Margin'] = df['Away Score'] - df['Home Score']
df['Winning Margin'] = df['Winning Margin'].abs()
df
```

```python
# Add a total points column

df['Total Points'] = df['Away Score'] + df['Home Score']
df
```

```python
# Finalized schedule as csv file
file_name = '2019-2020 NBA Schedule Final Results'

df.to_csv(file_name, sep='\t', encoding='utf-8')
```
